```
import pandas as pd
url = 'https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv'
fuel = pd.read_csv(url, error_bad_lines=False)
#QUESTION 3
fuel.groupby('fuel_type_code_pudl')['fuel_cost_per_unit_burned'].mean()

#QUESTION 4
fuel['fuel_mmbtu_per_unit'].std()
fuel['fuel_mmbtu_per_unit'].quantile(q=0.75)

#QUESTION 5
kurtos = pd.DataFrame(fuel['fuel_qty_burned']).kurtosis()[0]
print(kurtos)
skew = pd.DataFrame(fuel['fuel_qty_burned']).skew()[0]
print(skew)

#QUESTION 6
fuel.isnull().sum()
a = fuel['fuel_unit'].isnull().sum()
b = fuel['fuel_unit'].count()
print((a/b) * 100)

#QUESTION 8
fuel.corr(method='pearson')

# QUESTION 9 
fuel.groupby('report_year')['fuel_cost_per_unit_burned'].sum()

# QUESTION 10 
fuel.groupby('report_year')['fuel_cost_per_unit_delivered'].mean()
```
